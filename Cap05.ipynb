```python
class Pessoa:
    def __init__ (self, nome, idade):
        self.nome = nome
        self.idade = idade
    def faz_aniversario(self):
        self.idade += 1
    def __str__ (self):
        return f"{self.nome}, {self.idade} anos."

pessoa1 = Pessoa("Amanda", 35)
pessoa2= Pessoa("Renato", 32)

Aniversario = "Antes do Aniversario"
Depois = "Depois do Aniversario"
print(Aniversario)
print(pessoa1)
print(pessoa2)

pessoa1.faz_aniversario()
pessoa2.faz_aniversario()

print("-" * len(Aniversario))

print("Depois do Aniversario")
print(pessoa1)
print(pessoa2)
print("-" * len(Depois))
```

```python
class lampada: 
    def __init__(self): 
        self._estado = False 
    def ligar(self): 
        self._estado = True 
    def desligar(self): 
        self._estado = False
    def is_acesa(self):
        return self._estado

lampada = lampada()


print("A lampada está acesa?", lampada.is_acesa())
lampada.ligar()
print("A lampada está acesa?", lampada.is_acesa())
lampada.desligar()
print("A lampada está acesa?", lampada.is_acesa())
```

```python
class Animal:
    def fazer_som(self):
        print("Hakunamatata")
class Cachorro(Animal):
    def fazer_som(self):
        print("Au Au")
class Gato(Animal):
    def fazer_som(self):
        print("Miau")


animal = Animal()
cachorro = Cachorro()
gato = Gato()

print("O cachorro faz:", end =" ")
cachorro.fazer_som()
print("O gato faz:", end =" ")
gato.fazer_som()
print("E qualquer animal diz:", end=" ")
animal.fazer_som()
```

```python
def animais_falam(lista_animais):
    for animal in lista_animais:
        animal.fazer_som()
cachorro = Cachorro()
gato = Gato()
animal = Animal()
lista = [animal, cachorro , gato]
animais_falam(lista)
```
